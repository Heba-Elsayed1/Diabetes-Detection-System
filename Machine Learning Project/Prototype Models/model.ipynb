# Imports

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split  
from sklearn.metrics import  accuracy_score
from sklearn.feature_selection import mutual_info_classif
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

```

# Importing data

```python
data= pd.read_csv('data.csv')
data.head()
```

```python
data.info()
```

# Remove unnecessary columns

```python
data.drop("patient_number", inplace=True ,axis=1)
data
```

# checking for duplicate values

```python
data.duplicated().value_counts()
```

### no duplicates

# Encoding categorical data

```python
label_encoder= LabelEncoder()
data['gender']= label_encoder.fit_transform(data['gender'])
data['diabetes']= label_encoder.fit_transform(data['diabetes'])
data
```

# Extract features and target

```python
x= data.iloc[:,:-1]
y= data.iloc[:,-1]
```

```python
x
```

# Checking MI score

```python
discrete_features = x.dtypes == int
mi_scores = mutual_info_classif(x, y, discrete_features=discrete_features)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=x.columns)
mi_scores = mi_scores.sort_values(ascending=False)
mi_scores
```

# Splitting the dataset into train and test

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.15, random_state=0)
print(f"Training set size: {len(x_train)}")
# print(f"Cross-validation set size: {len(X_cv)}")
print(f"Testing set size: {len(x_test)}")
```

# Creating Logistic Classifier

```python
# Logistic Regression
log_reg = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier',LogisticRegression(max_iter=10000, random_state=42))
])
log_reg.fit(x_train, y_train)

# Evaluate on cross-validation set
y_log_pred = log_reg.predict(x_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_log_pred))
```

# Decision Tree Classifier

```python
# Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42,max_depth=3)
dt.fit(x_train, y_train)

# Evaluate on cross-validation set
y_dt_pred = dt.predict(x_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_dt_pred))
```

# Random Forest Classifier

```python
# Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

# Evaluate on cross-validation set
y_rf_pred = rf.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_rf_pred))
```
